```python
import pandas as pd
import numpy as np
import os
```

```python
files = [f for f in os.listdir('original')]
files = [f for f in files if f.endswith('.pickle')]
for file in files:
    df = pd.read_pickle(f'original\\{file}')
    df = df.sort_values(by='date').reset_index(drop = True)
    df['day_of_week'] = df['date'].dt.dayofweek
    # print(file)
    for index, row in df.iterrows():
        if index == 0:
            continue
        if (df.loc[index, 'date'] - df.loc[index-1, 'date']).days > 4 and (df.loc[index, 'date'] - df.loc[index-1, 'date']).days < 6 :
            print('Missing data in stock ', file.split('_')[0], 'from',df.loc[index-1, 'date'] ,'at', df.loc[index, 'date'])
```

```python
stocks = ['AMCR','ANET','BR','CEG','CHTR','DAL','DG','DOC','DOW','FOX','GM','HCA','KEYS','KMI','LIN','MRNA','MTCH','V','WBD']
files = [f for f in os.listdir('pickels') if f.split('_')[0] in stocks]
for file in files:
    df = pd.read_pickle(f'pickels\\{file}')
    df = df.sort_values(by='date').reset_index(drop = True)
    df['year'] = df['date'].dt.year
    for year in df['year'].unique():
        print(f'{file.split("_")[0]} -> {year}')
    print('---------------------------------')
```
